```julia
using Pkg
Pkg.activate(".")
using PsfUnmixing, Plots, LaTeXStrings, LinearAlgebra, Measures, DelimitedFiles
gr()
```

```julia
data_1 = readdlm("results/tsp_mse_projected_u=1.txt")
data_2 = readdlm("results/tsp_mse_projected_u=2.txt")
data_3 = readdlm("results/tsp_mse_projected_u=20.txt")

snr = data_1[1, :]

crb_1 = data_1[2, :]
mse_1 = data_1[3, :]

crb_2 = data_2[2, :]
mse_2 = data_2[3, :]

crb_3 = data_3[2, :]
mse_3 = data_3[3, :];
```

```julia

plt = plot(
	size = (1000, 400),
	yscale = :log10,
	legendfontsize = 15,
	xtickfontsize = 15,
	ytickfontsize = 15,
	legendposition = :topright,
)

plot!(
	plt,
	snr,
	mse_1,
	marker = (:square, 5),
	color = :darkorange,
	label = "MSE "*L"u = 1",
)
plot!(
	plt,
	snr,
	crb_1,
	linestyle = :dashdot,
	color = :darkorange,
	lw = 2,
	label = "CRB "*L"u = 1",
)
plot!(
	plt,
	snr,
	mse_2,
	marker = (:square, 5),
	color = :blue,
	label = "MSE "*L"u = 2",
)
plot!(
	plt,
	snr,
	crb_2,
	linestyle = :dashdot,
	color = :blue,
	lw = 2,
	label = "MSE "*L"u = 2",
)
plot!(
	plt,
	snr,
	mse_3,
	marker = (:square, 5),
	color = :red,
	label = "MSE "*L"u = 20",
)
plot!(
	plt,
	snr,
	crb_3,
	linestyle = :dashdot,
	color = :red,
	lw = 2,
	label = "CRB "*L"u = 20",
)

savefig(plt, "figures/icassp_mse_and_crb_vs_snr.pdf")
display(plt)
```
